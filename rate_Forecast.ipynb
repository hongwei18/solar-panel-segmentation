```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
from datetime import timedelta
from statsmodels.tsa.seasonal import STL
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# ---------------------------
# 1. Load and Preprocess Data
# ---------------------------
# Read data from Excel file and sheet "MED (Transformed)"
df = pd.read_excel("/content/Rate Forecast (5).xlsx", sheet_name="EUR NC (Transformed)")

# Filter for SCFI only (assumes a column "Rate Type")
df = df[df["Rate Type"] == "SCFI"].copy()

# Parse the "Period" column (assumed format "YYYY-WW") into a datetime.
# Here, we append "-1" to represent Monday of that week.
df["ds"] = pd.to_datetime(df["Period"] + "-1", format="%Y-%W-%w", errors="coerce")

# Rename the rate column to "y" (ensure the column name matches your Excel file)
df.rename(columns={"Value": "y"}, inplace=True)

# Drop rows with missing dates or rate values
df.dropna(subset=["ds", "y"], inplace=True)
df.sort_values("ds", inplace=True)
df.reset_index(drop=True, inplace=True)

print("Head of SCFI data after parsing:")
print(df.head())

# ---------------------------
# 2. STL Decomposition for Seasonal Adjustment
# ---------------------------
# The paper applied STL (LOESS) to remove trend and seasonality.
# Since the data is weekly and seasonality is yearly, we use period=52.
# We assume the series is long enough (the paper used 548 data points).
series = df.set_index("ds")["y"]

# Perform STL decomposition
stl = STL(series, period=52, robust=True)
result = stl.fit()
# Plot decomposition for reference (optional)
result.plot()
plt.tight_layout()
plt.show()

# Create a seasonally adjusted series. Here, we use the remainder (resid) plus the trend.
# (There are different approaches; one common method is to remove the seasonal component.)
y_adjusted = result.trend + result.resid
df["y_adj"] = y_adjusted.values

# Optionally, plot original vs adjusted series
plt.figure(figsize=(10, 4))
plt.plot(series.index, series, label="Original")
plt.plot(series.index, df["y_adj"], label="Seasonally Adjusted", linestyle='--')
plt.title("Original vs Seasonally Adjusted SCFI")
plt.xlabel("Date")
plt.ylabel("SCFI Rate")
plt.legend()
plt.show()

# ---------------------------
# 3. Prepare Data for LSTM
# ---------------------------
# Use the adjusted series for modeling.
data = df[["y_adj"]].values  # shape: (n_samples, 1)

# Scale the data to [0, 1]
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)

def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length])
    return np.array(X), np.array(y)

sequence_length = 120  # as specified in the paper
X, y_seq = create_sequences(scaled_data, sequence_length)

# Split data into training and testing sets using roughly 76% training (as in the paper)
split_index = int(0.76 * len(X))
X_train, X_test = X[:split_index], X[split_index:]
y_train, y_test = y_seq[:split_index], y_seq[split_index:]

print("Training samples:", X_train.shape, y_train.shape)
print("Testing samples:", X_test.shape, y_test.shape)

# ---------------------------
# 4. Build the LSTM Model
# ---------------------------
# According to the paper, the LSTM model uses one LSTM layer with 120 hidden units,
# no dropout (0.0), tanh activation, and sigmoid recurrent activation.
model = Sequential()
model.add(LSTM(units=120,
               activation='tanh',
               recurrent_activation='sigmoid',
               dropout=0.0,
               input_shape=(sequence_length, 1)))
model.add(Dense(1))  # output layer

model.compile(optimizer='adam', loss='mean_squared_error')
model.summary()

# ---------------------------
# 5. Train the Model
# ---------------------------
# Train for 200 epochs (as specified)
history = model.fit(X_train, y_train, epochs=200, batch_size=16,
                    validation_data=(X_test, y_test), verbose=1)

# Plot training history (optional)
plt.figure(figsize=(10,4))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title("Training History")
plt.xlabel("Epoch")
plt.ylabel("Loss (MSE)")
plt.legend()
plt.show()

# ---------------------------
# 6. Evaluate the Model
# ---------------------------
predictions = model.predict(X_test)
# Inverse transform predictions and actual values
predictions_inversed = scaler.inverse_transform(predictions)
y_test_inversed = scaler.inverse_transform(y_test)

rmse = math.sqrt(mean_squared_error(y_test_inversed, predictions_inversed))
print("Test RMSE:", rmse)

plt.figure(figsize=(10,6))
plt.plot(y_test_inversed, label="Actual SCFI")
plt.plot(predictions_inversed, label="Predicted SCFI")
plt.title("Test Set: Actual vs Predicted SCFI")
plt.xlabel("Time Step")
plt.ylabel("SCFI Rate")
plt.legend()
plt.show()

# ---------------------------
# 7. Forecast Future SCFI Values
# ---------------------------
future_steps = 26  # Forecast the next 10 weeks
last_sequence = scaled_data[-sequence_length:]
predicted_future = []
current_sequence = last_sequence.copy()

for _ in range(future_steps):
    # Reshape to (1, sequence_length, 1)
    current_sequence_reshaped = current_sequence.reshape(1, sequence_length, 1)
    next_pred = model.predict(current_sequence_reshaped)
    predicted_future.append(next_pred[0, 0])
    # Slide the window: remove the first value and append the new prediction
    current_sequence = np.append(current_sequence[1:], next_pred, axis=0)

predicted_future_inversed = scaler.inverse_transform(np.array(predicted_future).reshape(-1, 1))

# Generate future dates assuming weekly intervals
last_date = df["ds"].iloc[-1]
future_dates = [last_date + pd.DateOffset(weeks=i+1) for i in range(future_steps)]

forecast_df = pd.DataFrame({"ds": future_dates, "Predicted_SCFI": predicted_future_inversed.flatten()})
print("Future Forecast:")
print(forecast_df)

plt.figure(figsize=(10,6))
plt.plot(df["ds"], df["y"], label="Historical SCFI")
plt.plot(forecast_df["ds"], forecast_df["Predicted_SCFI"], label="Forecasted SCFI", marker='o')
plt.title("SCFI Forecast for Next 10 Weeks")
plt.xlabel("Date")
plt.ylabel("SCFI Rate")
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
from datetime import timedelta
from statsmodels.tsa.seasonal import STL
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# For residual analysis
import statsmodels.api as sm
from statsmodels.graphics.tsaplots import plot_acf

# ---------------------------
# 1. Load and Preprocess Data
# ---------------------------
df = pd.read_excel("/content/Rate Forecast (5).xlsx", sheet_name="EUR NC (Transformed)")

# Filter for SCFI only
df = df[df["Rate Type"] == "SCFI"].copy()

# Parse "Period" (YYYY-WW) -> Datetime (Monday of that ISO week)
df["ds"] = pd.to_datetime(df["Period"] + "-1", format="%Y-%W-%w", errors="coerce")

# Rename the numeric rate column to "y"
df.rename(columns={"Value": "y"}, inplace=True)

# Drop rows with missing values
df.dropna(subset=["ds", "y"], inplace=True)
df.sort_values("ds", inplace=True)
df.reset_index(drop=True, inplace=True)

print("Head of SCFI data after parsing:")
print(df.head())

# ---------------------------
# 2. STL Decomposition
# ---------------------------
series = df.set_index("ds")["y"]
# Decompose with period=52 for weekly data (yearly cycle)
stl = STL(series, period=52, robust=True)
result = stl.fit()

# (Optional) Plot STL decomposition
result.plot()
plt.tight_layout()
plt.show()

# Store the residual (stationarized series)
df["resid"] = result.resid.values

# ---------------------------
# 3. Prepare Data for LSTM (using residual only)
# ---------------------------
data = df[["resid"]].values  # shape: (n_samples, 1)

# Scale residuals to [0, 1]
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)

def create_sequences(dataset, seq_length):
    X, y = [], []
    for i in range(len(dataset) - seq_length):
        X.append(dataset[i:i+seq_length])
        y.append(dataset[i+seq_length])
    return np.array(X), np.array(y)

seq_length = 120
X, y_seq = create_sequences(scaled_data, seq_length)

# Train/Test split (76% train, 24% test)
split_index = int(0.76 * len(X))
X_train, X_test = X[:split_index], X[split_index:]
y_train, y_test = y_seq[:split_index], y_seq[split_index:]

print(f"Training samples: {X_train.shape}, {y_train.shape}")
print(f"Testing samples:  {X_test.shape}, {y_test.shape}")

# ---------------------------
# 4. Build LSTM Model
# ---------------------------
model = Sequential()
model.add(LSTM(
    units=120,
    activation='tanh',
    recurrent_activation='sigmoid',
    dropout=0.0,
    input_shape=(seq_length, 1)
))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error')
model.summary()

# ---------------------------
# 5. Train the Model
# ---------------------------
history = model.fit(
    X_train, y_train,
    epochs=200,
    batch_size=16,
    validation_data=(X_test, y_test),
    verbose=1
)

# (Optional) Plot training history
plt.figure(figsize=(10, 4))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title("Training History")
plt.xlabel("Epoch")
plt.ylabel("Loss (MSE)")
plt.legend()
plt.show()

# ---------------------------
# 6. Evaluate the Model
# ---------------------------
predictions_resid_scaled = model.predict(X_test)
# Inverse transform from scaled residual back to actual residual
predictions_resid = scaler.inverse_transform(predictions_resid_scaled)
y_test_resid = scaler.inverse_transform(y_test)

# Indices corresponding to test set
# (We offset by seq_length because we started creating sequences from the first 120 points)
test_indices = df.index[seq_length + split_index : seq_length + split_index + len(y_test)]

# Reconstruct predicted SCFI for the in-sample test set by adding
# the known trend + seasonal from STL
trend_test = result.trend.iloc[test_indices]
seasonal_test = result.seasonal.iloc[test_indices]

predicted_scfis_test = trend_test.values + seasonal_test.values + predictions_resid.flatten()
actual_scfis_test = df["y"].iloc[test_indices].values

# Compute RMSE in the original SCFI units
rmse_test = math.sqrt(mean_squared_error(actual_scfis_test, predicted_scfis_test))
print("Test RMSE (Reconstructed SCFI):", rmse_test)

# Plot in-sample test predictions vs actual
plt.figure(figsize=(10, 6))
plt.plot(df["ds"].iloc[test_indices], actual_scfis_test, label="Actual SCFI")
plt.plot(df["ds"].iloc[test_indices], predicted_scfis_test, label="Predicted SCFI (Trend+Seasonal+Residual)")
plt.title("Test Set: Actual vs. Reconstructed SCFI Predictions")
plt.xlabel("Date")
plt.ylabel("SCFI Rate")
plt.legend()
plt.show()

# ---------------------------
# 7. Forecast Future (Naive Approach for Trend+Seasonal)
# ---------------------------
# 7a. Forecast residual for the next 26 weeks
future_steps = 26
last_sequence = scaled_data[-seq_length:]
predicted_future_resid = []
current_seq = last_sequence.copy()

for _ in range(future_steps):
    current_seq_reshaped = current_seq.reshape(1, seq_length, 1)
    next_pred_scaled = model.predict(current_seq_reshaped)
    next_pred_resid = scaler.inverse_transform(next_pred_scaled)
    predicted_future_resid.append(next_pred_resid[0, 0])
    # Slide
    current_seq = np.append(current_seq[1:], next_pred_scaled, axis=0)

# 7b. Extend the STL components
# We'll do a naive approach:
#   - Keep the final trend constant
#   - Repeat the last 52 weeks of seasonal in a cycle
last_trend_val = result.trend.iloc[-1]
seasonal_vals = result.seasonal.iloc[-52:].values  # last 52 weeks

pred_future_trend = []
pred_future_seasonal = []
for i in range(future_steps):
    cyc_index = i % 52
    trend_val = last_trend_val  # naive: keep trend constant
    seasonal_val = seasonal_vals[cyc_index]
    pred_future_trend.append(trend_val)
    pred_future_seasonal.append(seasonal_val)

pred_future_trend = np.array(pred_future_trend)
pred_future_seasonal = np.array(pred_future_seasonal)
pred_future_resid = np.array(predicted_future_resid)

# Combine them
full_future_scfis = pred_future_trend + pred_future_seasonal + pred_future_resid

# Build a DataFrame with future dates
last_date = df["ds"].iloc[-1]
future_dates = [last_date + pd.DateOffset(weeks=i+1) for i in range(future_steps)]
forecast_df = pd.DataFrame({
    "ds": future_dates,
    "Predicted_SCFI": full_future_scfis
})

print("\nNaive Future Forecast (Trend+Seasonality+Residual):")
print(forecast_df)

# Plot historical + future forecast
plt.figure(figsize=(10,6))
plt.plot(df["ds"], df["y"], label="Historical SCFI")
plt.plot(forecast_df["ds"], forecast_df["Predicted_SCFI"], label="Forecasted SCFI", marker='o')
plt.title("SCFI Forecast for Next 26 Weeks (Naive Trend+Seasonal Extension)")
plt.xlabel("Date")
plt.ylabel("SCFI Rate")
plt.legend()
plt.show()

# ---------------------------
# 8. Additional Validation & Testing
# ---------------------------

# 8a. Multiple Error Metrics (MAE, MAPE, SMAPE)
from sklearn.metrics import mean_absolute_error
mae_test = mean_absolute_error(actual_scfis_test, predicted_scfis_test)
mape_test = np.mean(np.abs((actual_scfis_test - predicted_scfis_test) / actual_scfis_test)) * 100
smape_test = 100 * np.mean(2 * np.abs(actual_scfis_test - predicted_scfis_test) /
                           (np.abs(actual_scfis_test) + np.abs(predicted_scfis_test)))

print("\n--- Additional Test Metrics ---")
print("Test MAE (Reconstructed SCFI):", mae_test)
print("Test MAPE (Reconstructed SCFI):", mape_test, "%")
print("Test SMAPE (Reconstructed SCFI):", smape_test, "%")

# 8b. Naive Baseline Forecast (Test Set)
#    We'll define naive baseline as "SCFI[t] = SCFI[t-1]" for each test time step
naive_pred = [actual_scfis_test[0]]  # Start with the first test point
for i in range(1, len(actual_scfis_test)):
    # Forecast is the previous actual
    naive_pred.append(actual_scfis_test[i-1])

naive_rmse = math.sqrt(mean_squared_error(actual_scfis_test, naive_pred))
print("\nNaive Baseline RMSE (Test Set):", naive_rmse)

plt.figure(figsize=(10, 6))
plt.plot(df["ds"].iloc[test_indices], actual_scfis_test, label="Actual SCFI")
plt.plot(df["ds"].iloc[test_indices], naive_pred, label="Naive Forecast SCFI")
plt.title("Test Set: Actual vs. Naive Baseline Forecast")
plt.xlabel("Date")
plt.ylabel("SCFI Rate")
plt.legend()
plt.show()

# 8c. Residual Analysis (ACF Plot)
#    Check if there's autocorrelation left in the test residuals
test_residuals = actual_scfis_test - predicted_scfis_test
plt.figure(figsize=(10, 4))
plot_acf(test_residuals, lags=30)
plt.title("ACF of Test Set Residuals")
plt.show()

print("\nDone with additional validation steps!")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
from statsmodels.tsa.seasonal import STL
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from pmdarima import auto_arima
from sklearn.linear_model import LinearRegression

# ---------------------------
# 1. Data Loading & Preparation
# ---------------------------
# Load data with proper type handling
df = pd.read_excel("/content/Rate Forecast (5).xlsx",
                  sheet_name="EUR NC (Transformed)",
                  dtype={'Period': str, 'Value': float})

# Filter for SCFI only
df = df[df["Rate Type"] == "SCFI"].copy()

# ---------------------------
# 2. Robust Date Conversion
# ---------------------------
# Convert YYYY-WW to datetime (Mondays)
def convert_weekly_date(period_str):
    try:
        year, week = map(int, period_str.split('-'))
        date = pd.to_datetime(f"{year}-{week}-1", format="%Y-%W-%w")
        # Handle edge cases around year boundaries
        if date.year != year:
            date = pd.to_datetime(f"{year}-12-31") - pd.offsets.Week(weekday=6)
        return date
    except:
        return pd.NaT

df["ds"] = df["Period"].apply(convert_weekly_date)

# Drop invalid dates and sort
df = df.dropna(subset=["ds"])
df = df.sort_values("ds").reset_index(drop=True)

# Create complete weekly index
full_dates = pd.date_range(start=df["ds"].min(),
                          end=df["ds"].max(),
                          freq='W-MON')
df = df.set_index("ds").reindex(full_dates).rename_axis("ds").reset_index()

# Forward fill missing values (alternative: interpolate)
df["y"] = df["Value"].ffill()
df = df.dropna(subset=["y"]).reset_index(drop=True)

# ---------------------------
# 3. STL Decomposition
# ---------------------------
series = df.set_index("ds")["y"]
stl = STL(series, period=52, robust=True)
result = stl.fit()

# Store components
df["trend"] = result.trend.values
df["seasonal"] = result.seasonal.values
df["resid"] = result.resid.values

# ---------------------------
# 4. Trend & Seasonal Handling
# ---------------------------
# Fit trend model
X_trend = np.arange(len(df)).reshape(-1, 1)
trend_model = LinearRegression()
trend_model.fit(X_trend, df["trend"].values.reshape(-1, 1))

# Calculate 3-year seasonal average
seasonal_values = df["seasonal"].values[-52*3:]
seasonal_avg = seasonal_values.reshape(3, 52).mean(axis=0)

# ---------------------------
# 5. LSTM Model Setup
# ---------------------------
SEQ_LENGTH = 52
BATCH_SIZE = 32

# Prepare residuals
scaler = MinMaxScaler(feature_range=(0, 1))
scaler.fit(df[["resid"]].values)

def create_sequences(data, seq_length):
    scaled = scaler.transform(data)
    X, y = [], []
    for i in range(len(scaled)-seq_length):
        X.append(scaled[i:i+seq_length])
        y.append(scaled[i+seq_length])
    return np.array(X), np.array(y)

X, y = create_sequences(df[["resid"]].values, SEQ_LENGTH)

# Split data
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# Build model
model = Sequential([
    LSTM(64, input_shape=(SEQ_LENGTH, 1), return_sequences=True,
        dropout=0.2, recurrent_dropout=0.2),
    LSTM(32, dropout=0.2),
    Dense(1)
])

model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
             loss='mse')

# Train with early stopping
early_stop = EarlyStopping(monitor='val_loss', patience=30, restore_best_weights=True)
history = model.fit(X_train, y_train, epochs=300, batch_size=BATCH_SIZE,
                   validation_data=(X_test, y_test), callbacks=[early_stop])

# ---------------------------
# 6. Forecasting & Evaluation
# ---------------------------
def generate_forecast(steps=26):
    # Trend component
    last_idx = len(df)
    future_trend = trend_model.predict(
        np.arange(last_idx, last_idx+steps).reshape(-1, 1)
    ).flatten()

    # Seasonal component
    future_seasonal = np.tile(seasonal_avg, steps//52 + 1)[:steps]

    # Residual forecast
    last_resid = scaler.transform(df[["resid"]].values[-SEQ_LENGTH:])
    lstm_preds = []
    current_seq = last_resid.copy()

    for _ in range(steps):
        pred = model.predict(current_seq.reshape(1, SEQ_LENGTH, 1))
        lstm_preds.append(pred[0,0])
        current_seq = np.append(current_seq[1:], pred, axis=0)

    lstm_resid = scaler.inverse_transform(np.array(lstm_preds).reshape(-1, 1))

    # Combine components
    return future_trend + future_seasonal + lstm_resid.flatten()

# Generate predictions
future_steps = 26
future_dates = pd.date_range(start=df["ds"].iloc[-1] + pd.DateOffset(weeks=1),
                           periods=future_steps,
                           freq='W-MON')
forecast = generate_forecast(future_steps)

# Create forecast DataFrame
forecast_df = pd.DataFrame({
    "Date": future_dates,
    "Forecast": forecast
})

# Plot results
plt.figure(figsize=(12, 6))
plt.plot(df["ds"], df["y"], label="Historical")
plt.plot(forecast_df["Date"], forecast_df["Forecast"], label="26-Week Forecast", color='red')
plt.title("SCFI Weekly Forecast")
plt.xlabel("Date")
plt.ylabel("SCFI Rate")
plt.legend()
plt.grid(True)
plt.show()
```

```python
"""
Enhanced SCFI Rate Forecasting System
Description: Production-ready solution with advanced diagnostics
"""

#%% ---------------------------- IMPORTS ---------------------------- %%
import pandas as pd
import numpy as np
import logging
from prophet import Prophet
from prophet.diagnostics import cross_validation, performance_metrics
from prophet.plot import plot_components
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import xlsxwriter
import sys
import os
from datetime import datetime
from scipy.stats import iqr

# Configure matplotlib backend
matplotlib.use('Agg')

#%% ------------------------ CONFIGURATION -------------------------- %%
CONFIG = {
    "data": {
        "path": "/content/Rate Forecast (5).xlsx",
        "sheet": "MED (Transformed)",
        "rate_type": "SCFI",
        "date_format": "%G-%V",
        "covid_start": "2020-03-01",
        "covid_end": "2022-03-01",
        "min_data_days": 365
    },
    "model": {
        "forecast_periods": 26,
        "confidence_interval": 0.95,
        "seasonality": {
            "yearly": True,
            "weekly": False,
            "monthly": {
                "enable": True,
                "period": 30.5,
                "fourier_order": 8
            }
        },
        "hyperparams": {
            "changepoint_prior_scale": 0.2,
            "seasonality_prior_scale": 15,
            "holidays_prior_scale": 15,
            "mcmc_samples": 0,
            "growth": "linear"
        }
    },
    "output": {
        "excel_path": "/content/Forecast_Report.xlsx",
        "number_format": "#,##0.00",
        "chart_theme": {
            "actual_color": "#FF6B00",
            "forecast_color": "#0D6ABF",
            "fill_color": "#0D6ABF33",
            "grid_color": "#DDDDDD",
            "font_name": "Arial",
            "watermark": "CONFIDENTIAL"
        }
    }
}

#%% ------------------------ LOGGING SETUP -------------------------- %%
logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s",
    handlers=[
        logging.FileHandler("forecast.log"),
        logging.StreamHandler()
    ]
)
logger = logging.getLogger(__name__)

#%% ------------------------ DATA PIPELINE -------------------------- %%
class EnhancedDataProcessor:
    """Advanced data processing with feature engineering"""

    def __init__(self, config):
        self.config = config
        self.raw_data = None
        self.processed_data = None

    def load_data(self):
        """Load and validate dataset"""
        try:
            logger.info("Loading data from %s", self.config['data']['path'])
            self.raw_data = pd.read_excel(
                self.config['data']['path'],
                sheet_name=self.config['data']['sheet']
            )

            required_columns = ['Period', 'Rate Type', 'Value']
            missing = [col for col in required_columns
                      if col not in self.raw_data.columns]
            if missing:
                raise ValueError(f"Missing columns: {missing}")

            logger.info("Data loaded successfully. Rows: %d", len(self.raw_data))

        except Exception as e:
            logger.error("Data loading failed: %s", str(e))
            sys.exit(1)

    def preprocess(self):
        """Advanced preprocessing pipeline"""
        try:
            df = self.raw_data.copy()

            # Standardize columns
            df.columns = df.columns.str.strip().str.lower()
            df = df[df['rate type'] == self.config['data']['rate_type']]

            # Date conversion
            df['ds'] = pd.to_datetime(
                df['period'] + '-1',
                format=f"{self.config['data']['date_format']}-%u",
                errors='coerce'
            )

            # Clean invalid dates
            invalid_dates = df[df['ds'].isna()]
            if not invalid_dates.empty:
                logger.warning("Dropping %d invalid dates", len(invalid_dates))
                df = df.dropna(subset=['ds'])

            # Remove duplicates
            df = df.sort_values('ds').drop_duplicates('ds', keep='last')

            # Advanced outlier handling
            window_size = 12  # 12-week rolling window
            df['value_median'] = df['value'].rolling(window=window_size, center=True).median()
            df['value'] = np.where(
                abs(df['value'] - df['value_median']) > 2 * df['value'].std(),
                df['value_median'],
                df['value']
            )
            df = df.drop(columns=['value_median'])

            # Feature engineering
            df['covid_impact'] = df['ds'].apply(
                lambda x: 1 if pd.Timestamp(self.config['data']['covid_start']) <= x <= pd.Timestamp(self.config['data']['covid_end']) else 0
            )

            # Final structure
            self.processed_data = df[['ds', 'value', 'covid_impact']].rename(
                columns={'value': 'y'}
            ).sort_values('ds').reset_index(drop=True)

            # Data sufficiency check
            if (df['ds'].max() - df['ds'].min()).days < self.config['data']['min_data_days']:
                raise ValueError(f"Minimum {self.config['data']['min_data_days']} days of data required")

            logger.info("Enhanced preprocessing completed. Final rows: %d",
                      len(self.processed_data))

        except Exception as e:
            logger.error("Preprocessing failed: %s", str(e))
            sys.exit(1)

#%% ---------------------- MODELING FRAMEWORK ----------------------- %%
class EnhancedForecaster:
    """Optimized forecasting engine with diagnostics"""

    def __init__(self, config):
        self.config = config
        self.model = None
        self.forecast = None
        self.metrics = None

    def build_model(self):
        """Configure optimized Prophet model"""
        try:
            self.model = Prophet(
                interval_width=self.config['model']['confidence_interval'],
                yearly_seasonality=self.config['model']['seasonality']['yearly'],
                weekly_seasonality=self.config['model']['seasonality']['weekly'],
                changepoint_prior_scale=self.config['model']['hyperparams']['changepoint_prior_scale'],
                seasonality_prior_scale=self.config['model']['hyperparams']['seasonality_prior_scale'],
                holidays_prior_scale=self.config['model']['hyperparams']['holidays_prior_scale'],
                mcmc_samples=self.config['model']['hyperparams']['mcmc_samples'],
                growth=self.config['model']['hyperparams']['growth']
            )

            # Enhanced seasonality configuration
            if self.config['model']['seasonality']['monthly']['enable']:
                self.model.add_seasonality(
                    name='monthly',
                    period=self.config['model']['seasonality']['monthly']['period'],
                    fourier_order=self.config['model']['seasonality']['monthly']['fourier_order'],
                    prior_scale=15
                )

            # Add external regressors
            self.model.add_regressor('covid_impact')

            self.model.add_country_holidays(country_name='CN')
            logger.info("Optimized model configured successfully")

        except Exception as e:
            logger.error("Model initialization failed: %s", str(e))
            sys.exit(1)

    def forecast_rates(self, df):
        """Generate enhanced forecasts"""
        try:
            logger.info("Training optimized model")
            self.model.fit(df)

            # Create future dataframe
            future = self.model.make_future_dataframe(
                periods=self.config['model']['forecast_periods'],
                freq='W'
            )

            # Add future regressors
            future['covid_impact'] = future['ds'].apply(
                lambda x: 1 if pd.Timestamp(self.config['data']['covid_start']) <= x <= pd.Timestamp(self.config['data']['covid_end']) else 0
            )

            # Generate predictions
            self.forecast = self.model.predict(future)
            self.forecast = pd.merge(
                self.forecast,
                df[['ds', 'y']],
                on='ds',
                how='left'
            )
            logger.info("Enhanced forecast generated")

        except Exception as e:
            logger.error("Forecasting failed: %s", str(e))
            sys.exit(1)

    def validate_model(self, df):
        """Comprehensive model validation"""
        try:
            logger.info("Starting advanced validation")
            cutoffs = pd.date_range(
                start=df['ds'].min() + pd.Timedelta(days=180),
                end=df['ds'].max() - pd.Timedelta(days=90),
                freq='90D'
            )

            df_cv = cross_validation(
                self.model,
                horizon='90 days',
                cutoffs=cutoffs,
                parallel="processes"
            )

            self.metrics = performance_metrics(df_cv)

            # Baseline model comparison
            last_value = df['y'].iloc[-1]
            naive_mape = np.mean(
                np.abs((df_cv['y'] - last_value) / df_cv['y'])
            ) * 100
            logger.info(f"Naive Model MAPE: {naive_mape:.2f}%")
            logger.info(f"Prophet Model MAPE: {self.metrics['mape'].mean():.2f}%")

            logger.info("Validation Metrics:\n%s", self.metrics.describe())

        except Exception as e:
            logger.error("Validation failed: %s", str(e))
            self.metrics = pd.DataFrame()

#%% --------------------- REPORTING & OUTPUT ------------------------ %%
class DiagnosticReportGenerator:
    """Advanced reporting with diagnostics"""

    def __init__(self, config, forecast, metrics):
        self.config = config
        self.forecast = forecast
        self.metrics = metrics

    def create_excel_report(self):
        """Generate error-resistant Excel report"""
        try:
            logger.info("Creating enhanced Excel report")
            os.makedirs(os.path.dirname(self.config['output']['excel_path']), exist_ok=True)

            with pd.ExcelWriter(
                self.config['output']['excel_path'],
                engine='xlsxwriter',
                datetime_format='yyyy-mm-dd'
            ) as writer:
                # Forecast Sheet
                forecast_df = self.forecast[['ds', 'y', 'yhat', 'yhat_lower', 'yhat_upper']].copy()
                forecast_df.columns = ['Date', 'Actual', 'Forecast', 'Lower Bound', 'Upper Bound']
                forecast_df['Date'] = forecast_df['Date'].dt.strftime('%Y-%m-%d')

                # Safe error calculation
                forecast_df['Error (%)'] = ((forecast_df['Forecast'] - forecast_df['Actual']) /
                                          forecast_df['Actual'].replace(0, np.nan) * 100).round(2)

                forecast_df.to_excel(writer, sheet_name='Forecast', index=False)

                # Performance Metrics
                if not self.metrics.empty:
                    metrics_df = self.metrics.describe().T.reset_index()
                    metrics_df.columns = ['Metric', 'Count', 'Mean', 'Std', 'Min', '25%', '50%', '75%', 'Max']
                    metrics_df.to_excel(writer, sheet_name='Performance Metrics', index=False)

                # Documentation
                workbook = writer.book
                worksheet = workbook.add_worksheet('Documentation')
                content = [
                    ["Enhanced Forecast Report"],
                    ["Generated at", datetime.now().strftime('%Y-%m-%d %H:%M:%S')],
                    ["Model Configuration", str(self.config['model'])],
                    ["Key Enhancements", "1. Hyperparameter Tuning\n2. COVID Impact Regressor\n3. Advanced Validation"],
                    ["Data Source", self.config['data']['path']]
                ]
                for row_num, items in enumerate(content):
                    worksheet.write_row(row_num, 0, items)

            logger.info("Excel report generation successful")

        except Exception as e:
            logger.error("Report generation failed: %s", str(e))
            raise

    def create_diagnostic_plots(self):
        """Generate comprehensive visual diagnostics"""
        try:
            if self.forecast.empty:
                raise ValueError("No forecast data available for visualization")

            # Main forecast plot
            plt.style.use('seaborn-v0_8')
            fig, ax = plt.subplots(figsize=(14, 7))

            ax.plot(self.forecast['ds'], self.forecast['y'],
                   color=self.config['output']['chart_theme']['actual_color'],
                   label='Actual Rates')
            ax.plot(self.forecast['ds'], self.forecast['yhat'],
                   color=self.config['output']['chart_theme']['forecast_color'],
                   linestyle='--', label='Forecast')
            ax.fill_between(self.forecast['ds'], self.forecast['yhat_lower'],
                          self.forecast['yhat_upper'],
                          color=self.config['output']['chart_theme']['fill_color'],
                          label='95% CI')

            ax.set_title('Enhanced SCFI Rate Forecast', fontsize=16)
            ax.set_xlabel('Date', fontsize=12)
            ax.set_ylabel('Rate Value', fontsize=12)
            ax.grid(color=self.config['output']['chart_theme']['grid_color'], linestyle='--')
            ax.legend(loc='upper left')

            ax.xaxis.set_major_locator(mdates.MonthLocator(interval=2))
            ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
            fig.autofmt_xdate()

            plt.text(0.5, 0.5, self.config['output']['chart_theme']['watermark'],
                    fontsize=40, color='gray', alpha=0.2,
                    ha='center', va='center', rotation=30, transform=ax.transAxes)

            plt.tight_layout()
            plt.savefig('enhanced_forecast.png', dpi=300)
            plt.show()

            # Residual analysis
            self._plot_residuals()

            # Components analysis
            self._plot_components()

            logger.info("Diagnostic visuals generated")

        except Exception as e:
            logger.error("Visualization failed: %s", str(e))
            raise

    def _plot_residuals(self):
        """Plot prediction residuals"""
        residuals = self.forecast['y'] - self.forecast['yhat']

        plt.figure(figsize=(14, 6))
        plt.scatter(self.forecast['ds'], residuals, alpha=0.6)
        plt.axhline(0, color='red', linestyle='--')
        plt.title('Residual Analysis')
        plt.xlabel('Date')
        plt.ylabel('Prediction Error')
        plt.grid(True)
        plt.savefig('residuals.png', dpi=300)
        plt.show()

    def _plot_components(self):
        """Plot forecast components"""
        try:
            fig = plot_components(self.forecast.model, self.forecast)
            fig.set_size_inches(14, 10)
            plt.savefig('components.png', dpi=300)
            plt.show()
        except Exception as e:
            logger.warning("Component plot failed: %s", str(e))

#%% ------------------------ MAIN EXECUTION -------------------------- %%
if __name__ == "__main__":
    try:
        logger.info("=== Starting Enhanced Forecasting Pipeline ===")

        # Data Processing
        processor = EnhancedDataProcessor(CONFIG)
        processor.load_data()
        processor.preprocess()

        # Model Training & Forecasting
        forecaster = EnhancedForecaster(CONFIG)
        forecaster.build_model()
        forecaster.forecast_rates(processor.processed_data)
        forecaster.validate_model(processor.processed_data)

        # Generate Outputs
        reporter = DiagnosticReportGenerator(
            CONFIG,
            forecaster.forecast,
            forecaster.metrics
        )
        reporter.create_excel_report()
        reporter.create_diagnostic_plots()

        logger.info("=== Enhanced Pipeline Completed Successfully ===")

    except Exception as main_error:
        logger.critical("Pipeline failed: %s", str(main_error))
        sys.exit(1)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter, MonthLocator
from matplotlib import rcParams
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Set the font globally (optional, applies to the entire plot)
rcParams['font.family'] = 'Segoe UI'
rcParams['figure.dpi'] = 300  # Set high DPI for sharper rendering

# Step 1: Load the dataset
file_path = r"/content/Rate Forecast (2).xlsx"  # Replace with your file path
sheet_name = "MED (Transformed)"  # Adjust the sheet name if needed

dataset = pd.read_excel(file_path, sheet_name=sheet_name)

# Step 2: Inspect the dataset
print("Dataset preview:")
print(dataset.head())
print("Columns in dataset:", dataset.columns)

# Step 3: Clean column names
dataset.columns = dataset.columns.str.strip().str.lower()  # Standardize column names
print("Cleaned column names:", dataset.columns)

# Step 4: Filter by Rate Type 'NVD Index'
if 'rate type' not in dataset.columns:
    raise KeyError("The 'Rate Type' column is missing. Ensure it is included in the dataset.")

dataset = dataset[dataset['rate type'] == 'NVD Index']
print("Filtered dataset for 'Rate Type' = 'NVD Index':")
print(dataset.head())

# Step 5: Convert the 'Period' column to datetime
if 'period' not in dataset.columns:
    raise KeyError("The 'Period' column is missing. Ensure it is included in the dataset.")

# Convert 'Period' (e.g., "2024-01") to datetime
try:
    dataset['ds'] = pd.to_datetime(dataset['period'] + '-1', format='%G-%V-%u')  # ISO Week Year format
except Exception as e:
    print(f"Error converting 'Period' column: {e}")
    raise

# Drop rows with invalid dates
dataset = dataset.dropna(subset=['ds'])

# Step 6: Set the target column
if 'value' not in dataset.columns:
    raise KeyError("The 'Value' column is missing. Ensure it is included in the dataset.")

dataset['y'] = dataset['value']  # Set 'y' as the target column

# Step 7: Fit Exponential Smoothing (ETS) model with weekly seasonality
ets_model = ExponentialSmoothing(
    dataset['y'],
    trend='add',        # Additive trend
    seasonal='add',     # Additive weekly seasonality
    seasonal_periods=7, # Weekly seasonality
    initialization_method="estimated"
).fit()

# Step 8: Create future predictions for 6 months (approximately 26 weeks)
forecast_steps = 26
forecast = ets_model.forecast(steps=forecast_steps)

# Create a DataFrame for the forecasted future values
future_dates = pd.date_range(
    start=dataset['ds'].iloc[-1] + pd.Timedelta(weeks=1),
    periods=forecast_steps,
    freq='W'
)
forecast_df = pd.DataFrame({'ds': future_dates, 'yhat': forecast})

# Combine historical and forecast data for output
historical_forecast = pd.DataFrame({'ds': dataset['ds'], 'historical_forecast': ets_model.fittedvalues})
combined = pd.concat([historical_forecast, forecast_df.rename(columns={'yhat': 'forecast_value'})], ignore_index=True)

# Merge with the original dataset for actual values
final_df = pd.merge(combined, dataset[['ds', 'y']].rename(columns={'y': 'actual_value'}), on='ds', how='left')

# Add Period (yyyy-ww) for consistent formatting
final_df['Period'] = final_df['ds'].dt.strftime('%Y-%U')

# Step 9: Save the final DataFrame to Excel
output_file = "/content/Forecast Results.xlsx"  # Specify your output file path
final_df.to_excel(output_file, index=False)
print(f"Forecast results saved to {output_file}")

# Step 10: Optional Visualization
fig, ax = plt.subplots(figsize=(6.28, 5.16), dpi=300)

# Plot actual values
ax.plot(dataset['ds'], dataset['y'], '-', color="orange", label="Actual Value")

# Plot historical forecasts
ax.plot(historical_forecast['ds'], historical_forecast['historical_forecast'], ':', color="#0D6ABF", label="Historical Forecast")

# Plot future forecasts
ax.plot(forecast_df['ds'], forecast_df['yhat'], '-', color="red", label="Forecast (Future)")

# Customize the plot
ax.legend(loc="upper left", fontsize=8)
plt.show()
```

**SCFI (EUR NC)**

```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter, MonthLocator
from matplotlib import rcParams

# Set the font globally (optional, applies to the entire plot)
rcParams['font.family'] = 'Segoe UI'
rcParams['figure.dpi'] = 300  # Set high DPI for sharper rendering

# Define file path and sheet names
file_path = r"/content/Rate Forecast (2).xlsx"  # Replace with your file path
sheet_names = ["EUR NC (Transformed)"]  # Adjust the sheet names if needed

# Loop through each sheet
for sheet_name in sheet_names:
    print(f"Processing sheet: {sheet_name}")

    # Load the dataset
    dataset = pd.read_excel(file_path, sheet_name=sheet_name)

    # Step 1: Inspect the dataset
    print("Dataset preview:")
    print(dataset.head())
    print("Columns in dataset:", dataset.columns)

    # Step 2: Clean column names
    dataset.columns = dataset.columns.str.strip().str.lower()  # Standardize column names
    print("Cleaned column names:", dataset.columns)

    # Step 3: Filter by Rate Type 'SCFI'
    if 'rate type' not in dataset.columns:
        raise KeyError("The 'Rate Type' column is missing. Ensure it is added to the dataset.")

    dataset = dataset[dataset['rate type'] == 'SCFI']
    print(f"Filtered dataset for 'Rate Type' = 'SCFI' in sheet {sheet_name}:")
    print(dataset.head())

    # Step 4: Convert the 'Period' column to datetime
    if 'period' not in dataset.columns:
        raise KeyError("The 'Period' column is missing. Ensure it is added to the dataset.")

    # Convert 'Period' (e.g., "2024-01") to datetime
    try:
        dataset['ds'] = pd.to_datetime(dataset['period'] + '-1', format='%G-%V-%u')  # ISO Week Year format
    except Exception as e:
        print(f"Error converting 'Period' column: {e}")
        raise

    # Drop rows with invalid dates
    dataset = dataset.dropna(subset=['ds'])

    # Step 5: Set the target column for Prophet
    if 'value' not in dataset.columns:
        raise KeyError("The 'Value' column is missing. Ensure it is added to the dataset.")

    dataset['y'] = dataset['value']  # Set 'y' as the target column

    # Step 6: Initialize the Prophet model and add seasonality
    model = Prophet(
        yearly_seasonality=True,  # Default yearly seasonality
        weekly_seasonality=True,  # Default weekly seasonality
    )

    # Add custom monthly seasonality if applicable
    model.add_seasonality(
        name='monthly', period=30.5, fourier_order=5  # Adjust period/fourier_order based on data
    )

    # Fit the model
    model.fit(dataset)

    # Step 7: Create future predictions for 6 months (approximately 26 weeks)
    future = model.make_future_dataframe(periods=26, freq='W')  # Forecast for 26 weeks
    forecast = model.predict(future)

    # Step 8: Separate historical and future forecasts
    forecast_future = forecast[forecast['ds'] > dataset['ds'].max()]
    forecast_historical = forecast[forecast['ds'] <= dataset['ds'].max()]

    # Step 9: Add Period (yyyy-ww) to forecast
    forecast['period'] = forecast['ds'].dt.strftime('%Y-%U')

    # Combine Actual Values, Historical Forecast, and Future Forecast
    forecast_table = pd.merge(
        forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper', 'period']],
        dataset[['ds', 'y']].rename(columns={'y': 'actual_value'}),
        on='ds',
        how='left'
    )

    # Label historical forecasts explicitly
    forecast_table['historical_forecast'] = forecast_table.apply(
        lambda row: row['yhat'] if not pd.isna(row['actual_value']) else None,
        axis=1
    )

    # Rename columns
    forecast_table = forecast_table.rename(columns={
        'yhat': 'forecast_value',
        'yhat_lower': 'forecast_lower_bound',
        'yhat_upper': 'forecast_upper_bound'
    })

    # Step 10: Save the forecast table to Excel
    output_file = f"/content/Forecast Results - {sheet_name}.xlsx"  # Change the path if needed
    forecast_table.to_excel(output_file, index=False)
    print(f"Forecast table saved to {output_file}")

    # Optional: Visualize the results
    fig, ax = plt.subplots(figsize=(6.28, 5.16), dpi=300)  # Adjust size for Power BI container and set DPI

    # Plot historical forecast as a dotted blue line
    ax.plot(forecast_historical['ds'], forecast_historical['yhat'], ':', color="#0D6ABF", label="Historical Forecast")

    # Plot future forecast as a solid red line
    ax.plot(forecast_future['ds'], forecast_future['yhat'], '-', color="red", label="Forecast (Future)")

    # Plot observed (actual) data as a solid orange line
    ax.plot(dataset['ds'], dataset['y'], '-', color="orange", label="Actual Value")

    # Add uncertainty intervals for future forecast
    ax.fill_between(
        forecast_future['ds'],
        forecast_future['yhat_lower'],
        forecast_future['yhat_upper'],
        color='red',
        alpha=0.2,
        label="Uncertainty Interval"
    )

    # Add legend
    ax.legend(loc="upper left", fontsize=8)

    # Customize the plot
    plt.show()

```

**NVD Index (EUR NC)**

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter, MonthLocator
from matplotlib import rcParams
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Set the font globally (optional, applies to the entire plot)
rcParams['font.family'] = 'Segoe UI'
rcParams['figure.dpi'] = 300  # Set high DPI for sharper rendering

# Define file path and sheet names
file_path = r"/content/Rate Forecast Test.xlsx"  # Replace with your file path
sheet_names = ["EUR NC (Transformed)"]  # Add the sheet you want to process

# Loop through the sheet names
for sheet_name in sheet_names:
    print(f"Processing sheet: {sheet_name}")

    # Load the dataset
    dataset = pd.read_excel(file_path, sheet_name=sheet_name)

    # Step 1: Inspect the dataset
    print("Dataset preview:")
    print(dataset.head())
    print("Columns in dataset:", dataset.columns)

    # Step 2: Clean column names
    dataset.columns = dataset.columns.str.strip().str.lower()  # Standardize column names
    print("Cleaned column names:", dataset.columns)

    # Step 3: Filter by Rate Type 'NVD Index'
    if 'rate type' not in dataset.columns:
        raise KeyError("The 'Rate Type' column is missing. Ensure it is added to the dataset.")

    dataset = dataset[dataset['rate type'] == 'NVD Index']
    print(f"Filtered dataset for 'Rate Type' = 'NVD Index' in sheet {sheet_name}:")
    print(dataset.head())

    # Step 4: Convert the 'Period' column to datetime
    if 'period' not in dataset.columns:
        raise KeyError("The 'Period' column is missing. Ensure it is added to the dataset.")

    # Convert 'Period' (e.g., "2024-01") to datetime
    try:
        dataset['ds'] = pd.to_datetime(dataset['period'] + '-1', format='%G-%V-%u')  # ISO Week Year format
    except Exception as e:
        print(f"Error converting 'Period' column: {e}")
        raise

    # Drop rows with invalid dates
    dataset = dataset.dropna(subset=['ds'])

    # Step 5: Set the target column
    if 'value' not in dataset.columns:
        raise KeyError("The 'Value' column is missing. Ensure it is added to the dataset.")

    dataset['y'] = dataset['value']  # Set 'y' as the target column

    # Step 6: Handle outliers using percentile clipping
    q1, q3 = dataset['y'].quantile([0.05, 0.95])  # 5th and 95th percentiles
    dataset['y'] = dataset['y'].clip(lower=q1, upper=q3)

    # Step 7: Fit Exponential Smoothing (ETS) model with weekly seasonality
    ets_model = ExponentialSmoothing(
        dataset['y'],
        trend='add',        # Additive trend
        seasonal='add',     # Additive seasonality
        seasonal_periods=7, # Weekly seasonality
        initialization_method="estimated"
    ).fit()

    # Step 8: Create future predictions for 6 months (approximately 26 weeks)
    forecast_steps = 26
    forecast = ets_model.forecast(steps=forecast_steps)

    # Create a DataFrame for the forecasted future values
    future_dates = pd.date_range(
        start=dataset['ds'].iloc[-1] + pd.Timedelta(weeks=1),
        periods=forecast_steps,
        freq='W'
    )
    forecast_df = pd.DataFrame({'ds': future_dates, 'forecast_value': forecast})

    # Combine historical and forecast data into a unified table
    historical_forecast = pd.DataFrame({'ds': dataset['ds'], 'historical_forecast': ets_model.fittedvalues})
    historical_forecast['actual_value'] = dataset['y']

    combined_forecast = pd.concat([
        historical_forecast,
        forecast_df.assign(historical_forecast=None, actual_value=None)
    ], ignore_index=True)

    # Add Period (yyyy-ww) to forecast table
    combined_forecast['Period'] = combined_forecast['ds'].dt.strftime('%Y-%U')

    # Step 9: Save the forecast table to a new Excel sheet
    output_file = f"/content/Forecast Results - {sheet_name}.xlsx"
    combined_forecast.to_excel(output_file, index=False)
    print(f"Forecast table saved to {output_file}")

    # Optional: Visualize the results
    fig, ax = plt.subplots(figsize=(6.28, 5.16), dpi=300)

    # Plot observed values (actual data) as a solid orange line
    ax.plot(dataset['ds'], dataset['y'], '-', color="orange", label="Actual Value")

    # Plot modeled historical forecast (dotted line)
    ax.plot(dataset['ds'], ets_model.fittedvalues, ':', color="#0D6ABF", label="Historical Forecast")

    # Plot predicted future values (solid line)
    ax.plot(forecast_df['ds'], forecast_df['forecast_value'], '-', color="red", label="Forecast (Future)")

    # Add legend
    ax.legend(loc="upper left", fontsize=8)

    # Customize the plot
    plt.show()
```
